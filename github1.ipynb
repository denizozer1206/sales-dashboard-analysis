```python
import pandas as pd
df= pd.read_csv("Desktop/train.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["Postal Code"])
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Month"]= df["Order Date"].dt.to_period("M")
```

```python
df.groupby("Month")["Sales"].sum()
```

Sales peak observed in November and December across all years.
Consistent growth trend observed from 2015 to 2018.
Seasonal drop in February sales.

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

Technology category generated the highest revenue among all categories.
Furtinure and Office Supplies showed similar sales performance.
Technology products are the main revenue driver.

```python
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

West region generated the highest revenue.
South region showed the lowest sales performance.
East region has strong growth potential.
